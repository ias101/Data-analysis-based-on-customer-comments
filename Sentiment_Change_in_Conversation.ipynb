```python
import pandas as pd

result = pd.read_csv('conversations_56377143_vader.csv')
```

```python
result.head()
```

```python
result[['entity_user_mentions', 'id_str']] = result[['entity_user_mentions', 'id_str']].astype(str) 
```

```python
import numpy as np
```

```python
result['last_id'] = np.nan
```

```python
result['count_2'] = np.nan
```

```python
for index,info in result[['con_id_str_list2']].iterrows():
    new = list(filter(None, map(str.strip, info[0].strip('][').replace('', '').split(','))))
    if len(new) >= 1:
        result['count_2'][index] = 1
        result['last_id'][index] = new
    else:
        result['count_2'][index] = 0
```

```python
print(result[['last_id', 'count_2']])
```

```python
dropped = result[result['count_2']== 1]
```

```python
print(dropped)
```

```python
dropped['before_score'] = np.nan
```

```python
dropped['after_score'] = np.nan
```

```python
dropped['vader'] = np.nan
```

```python
for index,info in dropped[['full_conversation_vader']].iterrows():
    new = list(filter(None, map(str.strip, info[0].strip('][').replace('', '').split(','))))
    dropped['vader'][index] = new
    dropped['before_score'][index] = new[0]
    dropped['after_score'][index] = new[-1]
```

```python
# dropped = dropped.drop(3005)
```

```python
dropped = dropped.dropna()
```

```python
print(dropped)
```

```python
dropped['before_class'] = np.nan
```

```python
dropped['after_class'] = np.nan
```

```python
for index,info in dropped[['before_score']].iterrows():
    new = float(info[0].strip("'"))
    dropped['before_score'][index] = new
    if -1<= new < -0.25:
        dropped['before_class'][index] = 'Negative'
    elif -0.25<= new <0.25:
        dropped['before_class'][index] = 'Neutral'
    else:
        dropped['before_class'][index] = 'Positive'
```

```python
for index,info in dropped[['after_score']].iterrows():
    new = float(info[0].strip("'"))
    dropped['after_score'][index] = new
    if -1<= new < -0.25:
        dropped['after_class'][index] = 'Negative'
    elif -0.25<= new <0.25:
        dropped['after_class'][index] = 'Neutral'
    else:
        dropped['after_class'][index] = 'Positive'
```

```python
dropped
```

```python
dropped.to_csv('KLM_sentiment_change.csv')
```
